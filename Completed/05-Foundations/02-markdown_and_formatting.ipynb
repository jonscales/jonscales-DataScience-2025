## 4. LaTeX Math

You can render math using LaTeX syntax inside `$ ... $` for inline or `$$ ... $$` for block math.

Example:  
Inline: $E = mc^2$  
Block:
$$
\frac{\partial f}{\partial x} = 2x
$$

‚úÖ **Your Turn**: Write the quadratic formula in LaTeX.

```python
#Working with LaTeX   - the quadratic formula
%%markdown
#**The quadratic equation**
$ax^2 + bx + c = 0$

##**solve for x as follows**
$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a} $

<div align="center">
The quadratic formula in all its glory


$$
ax^2 + bx + c = 0
$$


$$
x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
$$

</div>
```

## 5. Mixing Code and Markdown

```python
# Example: create data in Python
numbers = [1, 2, 3, 4, 5]
squares = [n**2 for n in numbers]
squares
```

Now we explain results in Markdown:
- The list `numbers` contained 1‚Äì5.
- The list `squares` contains each number squared.
- Mixing narrative + results is why notebooks are so powerful.

‚úÖ **Your Turn**: Create a list of 3 numbers, calculate their cubes in code, then explain in Markdown.

```python
# list comprehension for next cell
numbers = [1,2,3]
cubes =[n**3 for n in numbers]
print(cubes)
```

```python
#mixed code & markdown


%%markdown
###Given any list of numbers (e.g. [1, 2, 3]), one can use list comprehension to create a second list by modifying
###each item in the initial list (e.g. cube each value in the first list)

##This will be demonstrated here:
###1. define a list

*numbers = [1,2,3]*

###2. define a 2nd list using list comprehension

*numbers = [1,2,3]*
*cubes = [n**3 for n in numbers]*

  ### for each value n in the list *numbers*, the value will be cubed,
$n^3$
$1^3 = 1 $ ;  $ 2^3 = 8 $ ; $ 3^3 = 27$

###4. display the new list of cubes using a print command

*print(cubes)*


###Results
  [1, 8, 27]






```
