## 1. Setup

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Line Plot

```python
x = np.linspace(1, 10, 100)
y = np.sin(x)


plt.plot(x, y, label="sin(x)")
plt.title("Line Plot Example")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.show()
```

## 3. Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 300 * np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis')
plt.title("Scatter Plot Example")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label="color scale")
plt.show()
```

## 4. Bar Plot

```python
categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]

plt.bar(categories, values)
plt.title("Bar Plot Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

## 5. Customizing Plots

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)", linestyle="--", color="blue")
plt.plot(x, y2, label="cos(x)", linestyle=":", color="red")
plt.title("Customized Line Plot")
plt.xlabel("x (radians)")
plt.ylabel("value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#I wanted to do all the trig functions so I tasked ChapGPT to generate the code
# I knew I could figure it out, but was tired of fixing all the asymptotic values

import numpy as np
import matplotlib.pyplot as plt

# x range
x = np.linspace(-np.pi/2, np.pi/2, 1000)

# base trig
sinx = np.sin(x)
cosx = np.cos(x)
tanx = np.tan(x)

# reciprocal functions
secx = 1 / cosx
cscx = 1 / sinx
cotx = cosx / sinx

# mask extreme values near asymptotes
for arr in [tanx, secx, cscx, cotx]:
    arr[np.abs(arr) > 10] = np.nan

plt.figure(figsize=(10, 6))

plt.plot(x, sinx, label='sin(x)', color='blue')
plt.plot(x, cosx, label='cos(x)', color='red')
plt.plot(x, tanx, label='tan(x)', color='purple', linestyle='--')
plt.plot(x, secx, label='sec(x)', color='green', linestyle=':')
plt.plot(x, cscx, label='csc(x)', color='orange', linestyle='-.')
plt.plot(x, cotx, label='cot(x)', color='brown', linestyle='--')

# zero axis
plt.axhline(0, color='black', linewidth=0.5)

# Ï€ tick labels
ticks = [-np.pi/2, -np.pi/4, 0, np.pi/4, np.pi/2]
labels = [r"$-\pi/2$", r"$-\pi/4$", "0", r"$\pi/4$", r"$\pi/2$"]
plt.xticks(ticks, labels)

plt.title("Trig Functions from -Ï€ to Ï€")
plt.xlabel("x (radians)")
plt.ylabel("value")

plt.ylim(-10, 10)
plt.grid(True)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
