## Q6. Plotting
Create a line plot of $y = x^2$ for $x$ values from 0 to 10.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10, 100)
y=x**2

plt.plot(x,y,label="$y = x^2$")
plt.xlabel("x")
plt.ylabel("$y (x^2)$")
plt.title("Line Plot of $y=x^2$")

plt.legend()
plt.grid(True)
plt.show()
```

## Q9. Timing
Compare the runtime of squaring numbers using a Python loop vs. NumPy. Use `%timeit` for each.

```python
import numpy as np

numbers = np.arange(1000)
#loop
%timeit for n in numbers: n**2
# list comprehension
%timeit [n**2 for n in numbers]
#numpy
%timeit numbers**2
```
