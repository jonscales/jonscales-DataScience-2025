# ðŸ§¼ Cleaning Dirty Data (Missing Values & Type Fixes)

## ðŸ”¹ LEARNING GOALS:
- Detect and count missing values (`NaN`)
- Fill or drop missing data
- Convert column data types safely
- Understand the difference between `NaN`, `None`, `""`, and type mismatches


### ðŸ§ª 1. Load a Messy Dataset

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", None],
    "Age": ["25", "thirty", 35, np.nan, "40"],
    "Signup Date": ["2022-01-01", "not a date", "2022/03/01", None, "April 5, 2022"],
    "Score": [95.5, None, 88.0, 92.5, ""]
}

df = pd.DataFrame(data)
df
```

### ðŸ§¯ 2. Detecting Missing or Broken Values

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

### ðŸ§¹ 3. Cleaning Strategy Options

```python
df.fillna({
    "Name": "Unknown",
    "Age": -1,
    "Signup Date": "1970-01-01",
    "Score": 0.0
})
```

```python
df.dropna()
```

### ðŸ§¬ 4. Data Type Fixes

```python
df.dtypes
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Score"] = pd.to_numeric(df["Score"], errors="coerce")
df["Signup Date"] = pd.to_datetime(df["Signup Date"], errors="coerce")
df.dtypes
```

### ðŸ©¹ 5. Impute (Fill In) Fixed Missing Values

```python
df.fillna({"Age": (df["Age"].median())}, inplace=True)
df.fillna({"Score":(df["Score"].mean())}, inplace=True)
```

```python
df.fillna({"Signup Date":(df["Signup Date"].min())}, inplace=True)
```

```python
df.fillna({"Name":"Unknown"}, inplace=True)
df.head()

```

### ðŸ¤“ 6. Cleaned Data Review

```python
df.info()
```

```python
df.describe(include="all")
```

```python
import pandas as pd
import numpy as np

moddata = {
    "Name": ["Alice", "Bob", "Charlie", "David", None, "Joe"],
    "Age": ["25", "thirty", 35, np.nan, "40", "adult"],
    "Signup Date": ["2022-01-01", "not a date", "2022/03/01", None, "April 5, 2022", "1/10/25"],
    "Score": [95.5, None, 88.0, 92.5, "", "A"],
    "PreReg":["Y","Y","N","X", " ",None ]
}

df = pd.DataFrame(moddata)
df
```

```python
# assess the mess
print(df.isnull().sum())
df.dtypes
#  Why can't two df.  operations be displayed in the same code block?  It seems only the last df. method command is displayed so I found using print() on the others will get them to be displayed
```

```python
# fix the mess
grades ={
    "A":95,"B":85,"C":75,"D":65,"F":55
}


df["Score"]=df["Score"].replace(grades)
df["Score"] = pd.to_numeric(df["Score"], errors="coerce")
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Score"] = pd.to_numeric(df["Score"], errors="coerce")
df["Signup Date"] = pd.to_datetime(df["Signup Date"], errors="coerce")
df["PreReg"] = df["PreReg"].replace({"X":"Y",None:"N", " ":"N"})
print(df.dtypes)
df
```

```python
df.fillna({"Age": (df["Age"].median())}, inplace=True)
df.fillna({"Score": 0 }, inplace=True)
df["Signup Date"] = df["Signup Date"].fillna(pd.Timestamp.today().normalize())
df.fillna({"Name":"J Doe"}, inplace=True)
df
```
